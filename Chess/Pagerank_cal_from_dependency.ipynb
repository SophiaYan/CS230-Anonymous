```r
# generate new 
# input the text file, construct graph,calculate pagerank
library("igraph")
#dd <- read.table("test.txt")
edges.list <- read.table("DependenceGraph.txt")
class.graph <- graph.data.frame(edges.list, directed=TRUE)
plot(class.graph,vertex.size = 4)
```

```r
class.page <- page_rank(class.graph, algo = "prpack", vids = V(class.graph),
  directed = TRUE, damping = 0.85, personalized = NULL, weights = NULL,
  options = NULL)
# find the top 10 pagerank
top10.rank <- sort(class.page$vector,decreasing = TRUE)[1:10];
as.numeric(names(top10.rank))
```
